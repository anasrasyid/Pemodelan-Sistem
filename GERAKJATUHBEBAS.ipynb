```
import matplotlib.pyplot as plt
import numpy as np
import math

TIMESTAMP = 0.01
HIGH = 10
GRAVITY = -9.8

def H_ANALYTICS(t):
  return HIGH + GRAVITY*t*t/2

def H_NUMERIC(n):
  v_numeric = [0]
  h_numeric = [HIGH]
  timefinal = 0
  while(n > 1):
    v_numeric.append(v_numeric[-1] + GRAVITY * TIMESTAMP)
    h_numeric.append(h_numeric[-1] + v_numeric[-1] * TIMESTAMP)
    if(h_numeric[-1] > 0):
      timefinal += TIMESTAMP
    n = n - 1
  return h_numeric, timefinal


time = np.arange(0.0, math.sqrt(2*(0 - HIGH) / GRAVITY), TIMESTAMP)
h_numeric, timeNumeric = H_NUMERIC(len(time))

#PLOTING AREA
plt.figure(figsize=(10,5), dpi=100)
plt.ylabel("High")
plt.xlabel("Time")
plt.plot(ylabel="HIGH", xlabel='TIME')
plt.plot(time, H_ANALYTICS(time), 'k',label='ANALYTICS')
plt.plot(time,h_numeric,'b',label='NUMERIC',linestyle='dashed')
plt.legend()
plt.show()

#Ploting Manual

plt.figure(figsize=(20,5), dpi=100)
plt.subplot(131)
plt.ylabel("High")
plt.xlabel("Time")
timeAnalytics = np.arange(0.0, math.sqrt(2*(0 - HIGH) / GRAVITY), 0.01)
plt.plot(timeAnalytics, H_ANALYTICS(timeAnalytics), 'k')
plt.subplot(132)
plt.plot(time,h_numeric,'b',linestyle='dashed')
plt.suptitle('ANALYTICS VS NUMERIC')
plt.show()

print("Final Time ANALYTICS : {}".format(timeAnalytics[-1]))
print("Final Time NUMERIC : {}".format(timeNumeric))
```
